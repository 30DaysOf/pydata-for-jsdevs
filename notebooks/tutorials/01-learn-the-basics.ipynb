# Conditionals
Explain conditionals in Python, including if, elif, and else statements. Show examples of using conditionals in Python programs.

```python
# Conditionals

# Conditionals in Python are used to execute different code blocks based on certain conditions.

# The if statement is used to execute code if a condition is true.
# The elif statement is used to check additional conditions if the previous conditions are false.
# The else statement is used to execute code if all the previous conditions are false.

# Here is an example of using if, elif, and else statements in Python:

x = 5

if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but less than or equal to 10")
else:
    print("x is less than or equal to 5")

# Output: x is less than or equal to 5

# You can also use logical operators to combine conditions.
# The and operator returns True if both conditions are true.
# The or operator returns True if at least one condition is true.
# The not operator returns the opposite of the condition.

# Here is an example of using logical operators in Python:

x = 5
y = 10

if x > 1 and y > 1:
    print("Both x and y are greater than 1")

if x > 1 or y > 1:
    print("Either x or y is greater than 1")

if not x > 10:
    print("x is not greater than 10")

# Output: Both x and y are greater than 1
#         Either x or y is greater than 1
#         x is not greater than 10
```

# Exceptions
Explain exceptions in Python, including try, except, and finally statements. Show examples of handling exceptions in Python programs.

```python
# Exceptions

# Exceptions in Python are used to handle errors and unexpected situations that may occur during program execution.

# The try statement is used to enclose code that may raise an exception.
# The except statement is used to handle the exception and provide an alternative code path.
# The finally statement is used to execute code regardless of whether an exception was raised or not.

# Here is an example of using try, except, and finally statements in Python:

try:
    x = 1 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("This code will always execute")

# Output: Cannot divide by zero
#         This code will always execute

# You can also catch multiple exceptions using a tuple of exception types.
# You can use the else statement to execute code if no exceptions were raised.

# Here is an example of using multiple except statements and an else statement in Python:

try:
    x = int("abc")
except ValueError:
    print("Cannot convert string to integer")
except TypeError:
    print("Cannot perform operation on different data types")
else:
    print("Conversion successful")
finally:
    print("This code will always execute")

# Output: Cannot convert string to integer
#         This code will always execute
```

# Lists
Explain lists in Python, including creating, accessing, and modifying lists. Show examples of using lists in Python programs.

```python
# Lists

# Lists in Python are used to store collections of items, such as numbers or strings.
# Lists are created using square brackets [] and items are separated by commas.

# Here is an example of creating a list in Python:

numbers = [1, 2, 3, 4, 5]
fruits = ["apple", "banana", "cherry"]

# You can access individual items in a list using their index.
# Indexing starts at 0 for the first item in the list.

# Here is an example of accessing items in a list in Python:

print(numbers[0]) # Output: 1
print(fruits[1]) # Output: banana

# You can also use negative indexing to access items from the end of the list.
# -1 refers to the last item in the list, -2 refers to the second last item, and so on.

# Here is an example of using negative indexing to access items in a list in Python:

print(numbers[-1]) # Output: 5
print(fruits[-2]) # Output: banana

# You can modify items in a list by assigning a new value to their index.

# Here is an example of modifying items in a list in Python:

numbers[0] = 10
fruits[1] = "orange"

print(numbers) # Output: [10, 2, 3, 4, 5]
print(fruits) # Output: ["apple", "orange", "cherry"]

# You can add items to a list using the append() method.
# The append() method adds the item to the end of the list.

# Here is an example of adding items to a list in Python:

numbers.append(6)
fruits.append("grape")

print(numbers) # Output: [10, 2, 3, 4, 5, 6]
print(fruits) # Output: ["apple", "orange", "cherry", "grape"]

# You can remove items from a list using the remove() method.
# The remove() method removes the first occurrence of the specified item.

# Here is an example of removing items from a list in Python:

numbers.remove(3)
fruits.remove("cherry")

print(numbers) # Output: [10, 2, 4, 5, 6]
print(fruits) # Output: ["apple", "orange", "grape"]

# You can check if an item is in a list using the in keyword.
# The in keyword returns True if the item is in the list, and False otherwise.

# Here is an example of checking if an item is in a list in Python:

if 4 in numbers:
    print("4 is in the list")
else:
    print("4 is not in the list")

if "banana" in fruits:
    print("banana is in the list")
else:
    print("banana is not in the list")

# Output: 4 is in the list
#         banana is not in the list
```

# Tuples
Explain tuples in Python, including creating, accessing, and modifying tuples. Show examples of using tuples in Python programs.

```python
# Tuples

# Tuples in Python are similar to lists, but they are immutable, meaning they cannot be modified once created.
# Tuples are created using parentheses () and items are separated by commas.

# Here is an example of creating a tuple in Python:

fruits = ("apple", "banana", "cherry")

# You can access individual items in a tuple using their index, just like with lists.

# Here is an example of accessing items in a tuple in Python:

print(fruits[0]) # Output: apple
print(fruits[1]) # Output: banana

# You can also use negative indexing to access items from the end of the tuple, just like with lists.

# Here is an example of using negative indexing to access items in a tuple in Python:

print(fruits[-1]) # Output: cherry
print(fruits[-2]) # Output: banana

# You cannot modify items in a tuple, since tuples are immutable.

# Here is an example of trying to modify items in a tuple in Python:

# fruits[0] = "orange" # This will raise a TypeError

# You can convert a list to a tuple using the tuple() function.

# Here is an example of converting a list to a tuple in Python:

numbers = [1, 2, 3, 4, 5]
numbers_tuple = tuple(numbers)

print(numbers_tuple) # Output: (1, 2, 3, 4, 5)

# Tuples are often used to return multiple values from a function.

# Here is an example of using a tuple to return multiple values from a function in Python:

def get_name_and_age():
    name = "John"
    age = 30
    return name, age

name, age = get_name_and_age()

print(name) # Output: John
print(age) # Output: 30
```

# Sets
Explain sets in Python, including creating, accessing, and modifying sets. Show examples of using sets in Python programs.

```python
# Sets

# Sets in Python are used to store collections of unique items, meaning that each item can only appear once in the set.
# Sets are created using curly braces {} or the set() function, and items are separated by commas.

# Here is an example of creating a set in Python:

fruits = {"apple", "banana", "cherry"}

# You can also create a set from a list using the set() function.

# Here is an example of creating a set from a list in Python:

numbers = [1, 2, 3, 4, 5, 5, 4, 3, 2, 1]
numbers_set = set(numbers)

print(numbers_set) # Output: {1, 2, 3, 4, 5}

# You can access individual items in a set using a for loop, since sets are unordered.

# Here is an example of accessing items in a set in Python:

for fruit in fruits:
    print(fruit)

# Output: apple
#         banana
#         cherry

# You cannot access individual items in a set using their index, since sets are unordered.

# Here is an example of trying to access items in a set using their index in Python:

# print(fruits[0]) # This will raise a TypeError

# You can add items to a set using the add() method.
# The add() method adds the item to the set if it is not already present.

# Here is an example of adding items to a set in Python:

fruits.add("orange")

print(fruits) # Output: {"apple", "banana", "cherry", "orange"}

# You can remove items from a set using the remove() method.
# The remove() method removes the specified item from the set.

# Here is an example of removing items from a set in Python:

fruits.remove("banana")

print(fruits) # Output: {"apple", "cherry", "orange"}

# You can check if an item is in a set using the in keyword.
# The in keyword returns True if the item is in the set, and False otherwise.

# Here is an example of checking if an item is in a set in Python:

if "apple" in fruits:
    print("apple is in the set")
else:
    print("apple is not in the set")

if "banana" in fruits:
    print("banana is in the set")
else:
    print("banana is not in the set")

# Output: apple is in the set
#         banana is not in the set
```

# Dictionaries
Explain dictionaries in Python, including creating, accessing, and modifying dictionaries. Show examples of using dictionaries in Python programs.

```python
# Dictionaries

# Dictionaries in Python are used to store collections of key-value pairs, where each key is unique.
# Dictionaries are created using curly braces {} and key-value pairs are separated by colons.

# Here is an example of creating a dictionary in Python:

person = {"name": "John", "age": 30, "city": "New York"}

# You can access individual values in a dictionary using their key.

# Here is an example of accessing values in a dictionary in Python:

print(person["name"]) # Output: John
print(person["age"]) # Output: 30

# You can also use the get() method to access values in a dictionary.
# The get() method returns None if the key is not found, or a default value if specified.

# Here is an example of using the get() method to access values in a dictionary in Python:

print(person.get("name")) # Output: John
print(person.get("gender")) # Output: None
print(person.get("gender", "unknown")) # Output: unknown

# You can modify values in a dictionary by assigning a new value to their key.

# Here is an example of modifying values in a dictionary in Python:

person["age"] = 40
person["city"] = "San Francisco"

print(person) # Output: {"name": "John", "age": 40, "city": "San Francisco"}

# You can add new key-value pairs to a dictionary by assigning a value to a new key.

# Here is an example of adding key-value pairs to a dictionary in Python:

person["gender"] = "male"

print(person) # Output: {"name": "John", "age": 40, "city": "San Francisco", "gender": "male"}

# You can remove key-value pairs from a dictionary using the del keyword.

# Here is an example of removing key-value pairs from a dictionary in Python:

del person["city"]

print(person) # Output: {"name": "John", "age": 40, "gender": "male"}

# You can check if a key is in a dictionary using the in keyword.
# The in keyword returns True if the key is in the dictionary, and False otherwise.

# Here is an example of checking if a key is in a dictionary in Python:

if "name" in person:
    print("name is in the dictionary")
else:
    print("name is not in the dictionary")

if "city" in person:
    print("city is in the dictionary")
else:
    print("city is not in the dictionary")

# Output: name is in the dictionary
#         city is not in the dictionary
```
