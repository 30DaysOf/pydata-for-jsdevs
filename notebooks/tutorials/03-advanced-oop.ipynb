# Introduction to OOP
Introduce OOP and its importance in programming. Explain how OOP helps in organizing and structuring code efficiently.

````python
# Define a class called Person
class Person:
    # Define the __init__ method with name and age as parameters
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # Define a method called say_hello
    def say_hello(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
    
# Create an instance of the Person class
person1 = Person("John", 25)

# Call the say_hello method on the person1 instance
person1.say_hello()
````

# Classes
Explain classes in Python, including creating classes, attributes, and methods. Show examples of using classes in Python programs.

````python
# Define a class called Car
class Car:
    # Define the __init__ method with make, model, and year as parameters
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    # Define a method called start_engine
    def start_engine(self):
        print(f"The {self.make} {self.model} from {self.year} is starting its engine.")
    
# Create an instance of the Car class
car1 = Car("Toyota", "Camry", 2021)

# Call the start_engine method on the car1 instance
car1.start_engine()
````

# Inheritance
Explain inheritance in Python, including creating subclasses and overriding methods. Show examples of using inheritance in Python programs.

````python
# Define a subclass called Student that inherits from the Person class
class Student(Person):
    # Define the __init__ method with name, age, and major as parameters
    def __init__(self, name, age, major):
        # Call the __init__ method of the Person class and pass name and age as arguments
        super().__init__(name, age)
        self.major = major
    
    # Override the say_hello method of the Person class
    def say_hello(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old. My major is {self.major}.")
    
# Create an instance of the Student class
student1 = Student("Jane", 20, "Computer Science")

# Call the say_hello method on the student1 instance
student1.say_hello()
````

# Methods
Explain methods in Python, including instance methods, class methods, and static methods. Show examples of using methods in Python programs.

````python
# Define a class called Rectangle
class Rectangle:
    # Define the __init__ method with width and height as parameters
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    # Define an instance method called area that calculates the area of the rectangle
    def area(self):
        return self.width * self.height
    
    # Define a class method called create_square that creates a square with equal width and height
    @classmethod
    def create_square(cls, side_length):
        return cls(side_length, side_length)
    
    # Define a static method called is_square that checks if a rectangle is a square
    @staticmethod
    def is_square(rectangle):
        return rectangle.width == rectangle.height

# Create an instance of the Rectangle class
rectangle1 = Rectangle(5, 10)

# Call the area method on the rectangle1 instance
print(rectangle1.area())

# Create a square using the create_square class method
square1 = Rectangle.create_square(7)

# Call the area method on the square1 instance
print(square1.area())

# Check if rectangle1 is a square using the is_square static method
print(Rectangle.is_square(rectangle1))

# Check if square1 is a square using the is_square static method
print(Rectangle.is_square(square1))
````
