# Most Viewed TED Talks

```python
# Create a table with the top 20 most liked TED talks by 
# sorting the values for the likes column in descending order

top_20_liked_talks = df.nlargest(20, 'likes')
display(top_20_liked_talks)
```

```python
# For each 'author' column value, calculate total and average values by view and like columns
# Then create a table with the top 20 authors by total number of views 
#  Show the number of talks given with column name "Talks Given"
#  Show the total number of views with column name "Total Views"
#  Show the average number of views per talk with column name "Avg Views/Talk"
#  Show the average number of likes per talk with column name "Avg Likes/Talk"
#  Show the highest number of views for a single talk with column name "Highest Viewed"
# Show the highest number of likes for a single talk with column name "Highest Liked"

# group the data by author and calculate the required statistics
author_stats = df.groupby('author').agg({'views': ['count', 'sum', 'mean'], 'likes': ['mean', 'max']})

# flatten the multi-level column index
author_stats.columns = ['_'.join(col).strip() for col in author_stats.columns.values]

# sort the data by total views and select the top 20 authors
top_20_authors = author_stats.sort_values(by='views_sum', ascending=False).head(20)

# rename the columns as required
top_20_authors = top_20_authors.rename(columns={'views_count': 'Talks Given', 'views_sum': 'Total Views', 'views_mean': 'Avg Views/Talk', 'likes_mean': 'Avg Likes/Talk', 'likes_max': 'Highest Liked'})

# add a new column for highest viewed talk
top_20_authors['Highest Viewed'] = df.loc[df['views'].idxmax(), 'views']

# display the top 20 authors table
display(top_20_authors)
```

```python
# For each 'author' column value, calculate total and average values by view and like columns
# Then create a table with the top 20 authors by average number of views per talk
#  Show the number of talks given with column name "Talks Given"
#  Show the total number of views with column name "Total Views"
#  Show the average number of views per talk with column name "Avg Views/Talk"
#  Show the average number of likes per talk with column name "Avg Likes/Talk"
#  Show the highest number of views for a single talk with column name "Highest Viewed"
# Show the highest number of likes for a single talk with column name "Highest Liked"

# group the data by author and calculate the required statistics
author_stats = df.groupby('author').agg({'views': ['count', 'sum', 'mean'], 'likes': ['mean', 'max']})

# flatten the multi-level column index
author_stats.columns = ['_'.join(col).strip() for col in author_stats.columns.values]

# calculate the average views per talk
author_stats['Avg Views/Talk'] = author_stats['views_sum'] / author_stats['views_count']

# sort the data by average views per talk and select the top 20 authors
top_20_authors = author_stats.sort_values(by='Avg Views/Talk', ascending=False).head(20)

# rename the columns as required
top_20_authors = top_20_authors.rename(columns={'views_count': 'Talks Given', 'views_sum': 'Total Views', 'likes_mean': 'Avg Likes/Talk', 'views_mean': 'Avg Views/Talk', 'likes_max': 'Highest Liked'})

# add a new column for highest viewed talk
top_20_authors['Highest Viewed'] = df.loc[df['views'].idxmax(), 'views']

# display the top 20 authors table
display(top_20_authors)
```

```python
# Identify the top 20 authors who had the most total views across all their talks
# Visualize the data in a stacked bar chart with the author name on the x-axis and the total views on the y-axis
# Use a color palette of your choice
# For each bar, show the stacked values for each talk given by that author that contributes to the total views

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from CSV file into a Pandas DataFrame
df = pd.read_csv('./../../data/kaggle/TED-talks.csv')

# group the data by author and calculate the total views
author_views = df.groupby('author')['views'].sum()

# sort the data by total views and select the top 20 authors
top_20_authors = author_views.sort_values(ascending=False).head(20)

# get the data for the top 20 authors
top_20_data = df[df['author'].isin(top_20_authors.index)]

# group the data by author and talk and calculate the views for each talk
talk_views = top_20_data.groupby(['author', 'title'])['views'].sum()

# unstack the data to create a stacked bar chart
talk_views.unstack().plot(kind='bar', stacked=True, figsize=(15, 10))

# set the title and axis labels
plt.title('Top 20 Authors by Total Views')
plt.xlabel('Author')
plt.ylabel('Total Views')

# show the plot
plt.show()

```

```python
#Create a table with the top 20 authors based on their total views across all their talk_views
#Show the table sorted by the total views in descending order
# For each row show the author, total number of talks given, total views across talks, highest view, lowest view and average view

# group the data by author and calculate the required statistics
author_stats = df.groupby('author').agg({'views': ['count', 'sum', 'max', 'min', 'mean']})

# flatten the multi-level column index
author_stats.columns = ['_'.join(col).strip() for col in author_stats.columns.values]

# sort the data by total views and select the top 20 authors
top_20_authors = author_stats.sort_values(by='views_sum', ascending=False).head(20)

# rename the columns as required
top_20_authors = top_20_authors.rename(columns={'views_count': 'Total Talks Given', 'views_sum': 'Total Views', 'views_max': 'Highest View', 'views_min': 'Lowest View', 'views_mean': 'Average View'})

# display the top 20 authors table
display(top_20_authors)
```

```python
# Identify the top 20 authors who had the most total views across all their talks
# Visualize the data in a stacked bar chart with the author name on the x-axis and the total views on the y-axis
# Write the x-axis labels vertically - for each author show their name and the total number of talks given
# Use a color palette of your choice
# For each bar, show the stacked values for each talk given by that author that contributes to the total views
# Hide the legend
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from CSV file into a Pandas DataFrame
df = pd.read_csv('./../../data/kaggle/TED-talks.csv')

# group the data by author and calculate the total views
author_views = df.groupby('author')['views'].sum()

# sort the data by total views and select the top 20 authors
top_20_authors = author_views.sort_values(ascending=False).head(20)

# get the data for the top 20 authors
top_20_data = df[df['author'].isin(top_20_authors.index)]

# group the data by author and talk and calculate the views for each talk
talk_views = top_20_data.groupby(['author', 'title'])['views'].sum()

# unstack the data to create a stacked bar chart
ax = talk_views.unstack().plot(kind='bar', stacked=True, figsize=(15, 10))

# set the title and axis labels
plt.title('Top 20 Authors by Total Views')
plt.xlabel('Author')
plt.ylabel('Total Views')

# write the x-axis labels vertically
plt.xticks(rotation=90)

# write the x-axis labels vertically - for each author show their name and the total number of talks given
labels = []
for author in top_20_authors.index:
    talks_given = df[df['author'] == author]['title'].nunique()
    label = f'{author} - {talks_given}'
    labels.append(label)
ax.set_xticklabels(labels)

# Use a color palette of your choice
sns.set_palette('husl')

# Hide the legend
ax.legend().set_visible(False)

# show the plot
plt.show()
```
