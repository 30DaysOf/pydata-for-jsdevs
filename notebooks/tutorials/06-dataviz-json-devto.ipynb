# Load Data from JSON File
Use Pandas to read the data from the JSON file into a DataFrame.

```python
# Load data from 2 JSON files into the same dataframe - both files have the same structure
# The JSON files are located in the data folder and are named as follows:
# 1. ./../../data/misc/devto-azure-p1-1000.json
# 2. ./../../data/misc/devto-azure-p2-1000.json

# Import necessary libraries
import pandas as pd

# Load data from 2 JSON files into the same dataframe
df1 = pd.read_json('./../../data/misc/devto-azure-p1-1000.json')
df2 = pd.read_json('./../../data/misc/devto-azure-p2-1000.json')
df = pd.concat([df1, df2], ignore_index=True)

# Display the first 5 rows of the dataframe
df.head()


```

```python

# List all the columns in the dataframe with their types
df.info()

```

```python
# Create a table of the unique user values and the number of times they appear in the dataframe
# Sort the table in descending order of the number of times the user appears in the dataframe
# Show all the rows in the table

user_counts = df['user'].value_counts().sort_values(ascending=False)
user_counts.head()

# Create a new dataframe with only the 'name', 'username', and 'count' columns
user_table = pd.DataFrame({'name': user_counts.index.str['name'], 'username': user_counts.index.str['username'], 'count': user_counts.values})

# Display the new dataframe
user_table.head()

```

```python
# Repeat the same process but this get data for the top 25 users in terms of total count
# And visualize as a bar chart with the user names on the x-axis and the number of times they appear in the dataframe on the y-axis
# Create a table of the unique user values and the number of times they appear in the dataframe
# Sort the table in descending order of the number of times the user appears in the dataframe
# Show only the top 25 rows in the table

user_counts = df['user'].value_counts().sort_values(ascending=False).head(25)

# Create a new dataframe with only the 'name', 'username', and 'count' columns
user_table = pd.DataFrame({'name': user_counts.index.str['name'], 'username': user_counts.index.str['username'], 'count': user_counts.values})

# Visualize the data as a bar chart
user_table.plot(x='username', y='count', kind='bar', figsize=(10, 6), title='Top 25 Users by Count')
plt.xlabel('Usernames')
plt.ylabel('Count')
plt.show()
```

```python
# Create a table of the unique users and the number of times they appear in the dataframe
# Sort the table in descending order of the number of times the user appears in the dataframe
# Show only the top 25 rows in the table
user_counts = df['user'].value_counts().sort_values(ascending=False).head(25)

# Create a new dataframe with only the 'name', 'username', and 'count' columns
user_table = pd.DataFrame({'name': user_counts.index.str['name'], 'username': user_counts.index.str['username'], 'count': user_counts.values})

# Display the new dataframe
user_table
```

```python
# Create a table of the unique users and the number of times they appear in the dataframe
# Sort the table in descending order of the number of times the user appears in the dataframe
# Show only the top 50 rows in the table
user_counts = df['user'].value_counts().sort_values(ascending=False).head(50)

# Create a new dataframe with only the 'name', 'username', and 'count' columns
user_table = pd.DataFrame({'name': user_counts.index.str['name'], 'username': user_counts.index.str['username'], 'count': user_counts.values})

# Add 3 new columns called #articles, #reactions, and #comments
# For each row in the dataframe, get the user's username and use it to filter the original dataframe
# Then set the value of the #articles column to the number of articles by that user
# Then set the value of the #reactions column to the sume of the 'public_reactions_count' column for all articles by that user
# Then set the value of the #comments column to the sume of the 'comments_count' column for all articles by that user
user_table['#articles'] = user_table['username'].apply(lambda x: df[df['user'].apply(lambda y: y['username'] == x)].shape[0])
user_table['#reactions'] = user_table['username'].apply(lambda x: df[df['user'].apply(lambda y: y['username'] == x)]['public_reactions_count'].sum())
user_table['#comments'] = user_table['username'].apply(lambda x: df[df['user'].apply(lambda y: y['username'] == x)]['comments_count'].sum())

# Display the new dataframe
user_table
```

```python
# Sort the above table in descending order of the #reactions column
# Show only the top 25 rows in the table
user_table.sort_values(by='#reactions', ascending=False).head(25)
```

```python
# Sort the user_table in descending order of the #comments column
# Show only the top 25 rows in the table
user_table.sort_values(by='#comments', ascending=False).head(25)
```

```python
# Visualize the data as a bar chart
# Show only the top 25 rows in the table
user_table.sort_values(by='#articles', ascending=False).head(25).plot(x='username', y='#articles', kind='bar', figsize=(10, 6), title='Top 25 Users by Article Count')
plt.xlabel('Usernames')
plt.ylabel('Articles Count')
plt.show()
```

```python
# Visualize the data as a bar chart this time by the #reactions column
# Show only the top 25 rows in the table
user_table.sort_values(by='#reactions', ascending=False).head(25).plot(x='username', y='#reactions', kind='bar', figsize=(10, 6), title='Top 25 Users by Reaction Count')
plt.xlabel('Usernames')
plt.ylabel('Reactions Count')
plt.show()
```

```python
# Create a new tags table that contains one row for each unique tag found in the tag_list array for each article
# The new table should have 2 columns: 'tag' and 'count' 
# The 'tag' column should contain the tag name
# The 'count' column should contain the number of times the tag appears in the tag_list array for all articles
# Sort the table in descending order of the number of times the tag appears in the tag_list array
# Show only the top 25 rows in the table


tags = df['tag_list'].explode().value_counts().sort_values(ascending=False).head(25)
tags_table = pd.DataFrame({'tag': tags.index, 'count': tags.values})
tags_table


```

```python
# Add a new column to the tags_table called 'authors' 
# Add a new column to the tags_table called '#authors'
# For each row in the tags_table, get the tag name and use it to filter the original dataframe
# Find all records where the tag appears in the tag_list column
# Then get a list of all unique users who are in the user column for all those records
# Then set the value of the 'authors' column to the list of unique users
# Then set the value of the '#authors' column to the number of unique users

tags_table['authors'] = tags_table['tag'].apply(lambda x: df[df['tag_list'].apply(lambda y: x in y)]['user'].apply(lambda z: z['username']).unique())
tags_table['#authors'] = tags_table['authors'].apply(lambda x: len(x))
tags_table
```

```python
# Visualize the tag count vs #authors as a scatter plot
# The x-axis should be the #authors column
# The y-axis should be the count column
# The size of each point should be the count column
# The color of each point should be the count column
# The title should be 'Tag Count vs #Authors'
# The x-axis label should be '#Authors'
# The y-axis label should be 'Tag Count'
# The point should have the tag name as the label
# Use the 'plasma' colormap
# Set the figure size to (10, 6)
# Show the plot
# Use a smaller font size for the labels

tags_table.plot.scatter(x='#authors', y='count', s='count', c='count', title='Tag Count vs #Authors', colormap='plasma', figsize=(10, 6))
plt.xlabel('#Authors', fontsize=8)
plt.ylabel('Tag Count', fontsize=8)
for i, txt in enumerate(tags_table['tag']):
    plt.annotate(txt, (tags_table['#authors'][i], tags_table['count'][i]), fontsize=8)
plt.show()
```
